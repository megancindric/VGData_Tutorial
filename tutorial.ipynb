## SORTING VALUES

```python
sales.head()
```

Can sort by a single column value, defaulting to an ascending order
To specify a different order, we can set ascending to False

```python
sales.sort_values("Name")
sales.sort_values("Name", ascending = False)
```

Since the sort_values() method results in a dataframe, we can chain additional methods like .head() or .tail()

```python
sales.sort_values("Platform").head()
```

## SORTING BY MULTIPLE COLUMNS

We can further specify columns to sort by by providing a list [] parameter.  Pandas will sort by first parameter, then within that sorting sort AGAIN by second parameter

To specify ascending/descending order, simply pass in second parameter.  NOTE: THIS APPLIES TO BOTH SORTING PARAMETERS!!

```python
sales.sort_values(["Platform", "Global_Sales"])
sales.sort_values(["Platform", "Global_Sales"], ascending = False)
```

## DIFFERENT SORT ORDERS FOR EACH SORT PARAMETER

By passing a list into the ASCENDING parameter, we can specify sort order for each specific sorting of data

Sorting Platform in ascending, Global Sales in descending:

```python
sales.sort_values(["Platform", "Global_Sales"], ascending = [True, False])
```

## MODIFYING THE ORIGINAL DATA STRUCTURE

By passing in the parameter inplace = True we modify original dataframe

```python
sales.sort_values(["Platform", "Global_Sales"], ascending = [True, False], inplace = True)
sales
```

## OPTIMIZATIONS

There are additional methods to provide useful info about our data set, and correct errors in our code

.info() will provide information about each column, it's data type, and how many non-null values are present
Converts numbers to float64 to account for null values 

```python
sales.info()
```

Note that data set has some null values!  We would want to remove those values to correct inaccuracies in data, and to bring down size of data set

We can use .dropna() to remove those values

Has a parameter: subset that takes in a list.  Pass in all column values, Pandas will check for null values, and remove those rows

```python
sales.dropna(subset = ["Year"])
sales.dropna(subset = ["Year"], inplace = True)
sales.head()
```

Now that null values are removed, we can convert float64 to integer values!

First, we will look at the series for Year

Then we will convert using .astype(int)

Once we check data is accurate, we can reassign that series with the converted output

```python
sales["Year"].astype(int)
sales["Year"] = sales["Year"].astype(int)
sales.head()
```

Behind the scenes, will keep one copy of each unique values, and all others will reference that single unique value

The lower the number of unique values, the higher chance it's a good time to use a category

To find number of unique values, can use .nunique()

```python
sales.nunique()
```

## Filtering

We can filter our dataframe to return a more specific subset of that data, based on a condition

First, will target specific column, then compare to a specific value

This will return a boolean series - True means we've found a match!

Can then specify this query WITHIN a query of our original dataframe.  Pandas will only return and display True values

To shorten syntax, can assign our filter to a variable, and plug that into our dataframe

```python
sales["Publisher"] == "Square Enix"
sales[sales["Publisher"] == "Square Enix"]

#Alternatively
square_enix = sales["Publisher"] == "Square Enix"
sales[square_enix]
```

Expanding on this example, we can filter dataframe to just find games with Action genre

```python
action_games = sales["Genre"] == "Action"
sales[action_games]
```

## AND Operator

Let's say we want to combine two filters

We can utilize an AND filter for this!

In Pandas, this is represented by &

Since the past two filters were saved in variables, we can use those variables in the comparison!

```python
sales[square_enix & action_games]
```

## OR operator

In Pandas, this is represented with |

```python
sales[square_enix | action_games]
```

## Querying By Numeric Comparison

Can utilize < and > to filter based on numeric values in a column

Can also use this to check between a RANGE of values

In Pandas, we can utilize .between() method

.between() takes two parameters - min and max values

Will INCLUDE these values as well!

```python
over_ten_mil = sales["Global_Sales"] > 10
sales[over_ten_mil]
```

## Group By

Say for example we want to find total global sales by each genre grouping

First, we can group columns by matching values, THEN sum each independently

Can utilize .groupby("Column_name")

## Melting

A different way for us to structure similar pieces of data

We can "group" similar columns together to get a more specific snapshot

The column headers of NA_SALES, EU_SALES, etc are essentially a grouping of region and sales

We can take this data and melt it into a "pivot table" using .melt()

Specify id_Vars for each column label you'd like to pull in

Specify value_vars for each column value you'd like to pull in (Will be parsed into separate row for each value)

Specify var_name for the column name corresponding to value_vars

Specify value_name for the column name corresponding to the VALUE from value_vars

```python
sales.reset_index(inplace=True)

# Region Sales
# NA     .15
# EU     .10
# JP     .05
```

```python
sales.melt(
    id_vars= ["Name", "Publisher"],
    value_vars = ["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"],
    var_name = "Region",
    value_name = "Sales"
    )
```
