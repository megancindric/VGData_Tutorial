## Working with a .json dataset

Using a dataset of all 480 Animal Crossing villagers

-Find all villagers that were in both New Horizons (NH) and New Leaf (NL)

-Determine the "most popular" kind of villager, based on number of unique villagers and frequency of appearances

```python
import pandas as pd
```

Reading the JSON file, and using the unique name of each villager as the index column

```python
villagers = pd.read_json('animal-crossing-villagers.json')
villagers.set_index("name", inplace=True)
villagers.head()
```

Grouping the table together by species

```python
villagers.sort_values(by="species", inplace=True)
villagers.head()
```

Counts of each type of animal species (METRIC #1) stored in variable species_count

```python
species_count = villagers["species"].value_counts()
species_count.head()
```

Creating a new column (appearance_count) representing the length of each list in the "Appearance" column

```python
villagers["appearances"].str.len()
villagers["appearances_count"] = villagers["appearances"].str.len()
villagers["appearances_count"]
```

Now we can utilize this column to group animals by villagers, and sum the related appearance_counts to see the unique number of appearances each species of animal has made

```python
appearances_by_species = villagers.groupby("species")["appearances_count"].sum().sort_values(ascending=False)
appearances_by_species.head()
```

```python
ratio_df = appearances_by_species / species_count
ratio_df.sort_values(ascending=False).head(10)
```
